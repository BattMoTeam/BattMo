
# <span style="color:rgb(213,80,0)">Equivalent Circuit Model</span>

## Load data set

```matlab
jsonstruct = parseBattmoJson('exampleECM.json');
inputparams = EquivalentCircuitModelInputParams(jsonstruct);
```

## Setup model

```matlab
model = EquivalentCircuitModel(inputparams);
```

## run Simulation

```matlab
[t, U, I] = model.solve();
```

## Plotting

```matlab
tiledlayout(1, 2);
nexttile
plot(t, U)
nexttile
plot(t, I)
```
